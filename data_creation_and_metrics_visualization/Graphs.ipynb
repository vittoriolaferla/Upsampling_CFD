```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


# Defining 5 different data points for mass coefficient and PSNR
mass_coefficient_1 = 0
mass_coefficient_2 = 2
mass_coefficient_3 = 5
mass_coefficient_4 = 8
mass_coefficient_5 = 10

psnr_1 = 22.48
psnr_2 = 20.86
psnr_3 = 20.33
psnr_4 = 20.34
psnr_5 = 20.29

# Storing them in lists for plotting
mass_coefficients = [mass_coefficient_1, mass_coefficient_2, mass_coefficient_3, mass_coefficient_4, mass_coefficient_5]
psnrs = [psnr_1, psnr_2, psnr_3, psnr_4, psnr_5]

# Fitting a linear regression line
coefficients = np.polyfit(mass_coefficients, psnrs, 1)
linear_fit = np.poly1d(coefficients)

# Generating x values for plotting the line
x_fit = np.linspace(min(mass_coefficients), max(mass_coefficients), 100)
y_fit = linear_fit(x_fit)

# Plotting the data points and linear regression line with fixed dimensions
plt.figure(figsize=(5, 3))  # Adjust width and height as needed
plt.scatter(mass_coefficients, psnrs, color='red', label='DM Results')
plt.title('Mass Coefficient vs PSNR')
plt.xlabel('Mass Coefficient')
plt.ylabel('PSNR')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
